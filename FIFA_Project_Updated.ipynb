## Task 1: Complete Data Analysis Report

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('players_20.csv')
df.head()
```

### Basic Info

```
df.info()
```

### Summary Statistics

```
df.describe()
```

## Task 2: Explore football skills and cluster players

```
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

features = df[['Crossing','Finishing','Dribbling','ShortPassing','LongPassing','BallControl','SprintSpeed','Acceleration','Strength','Agility']].dropna()
scaler = StandardScaler()
X = scaler.fit_transform(features)

kmeans = KMeans(n_clusters=5, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)
df[['Cluster']].head()
```

## Task 3: Exploratory Questions

### Top 10 Countries with Most Players

```
top_countries = df['Nationality'].value_counts().head(10)
print(top_countries)
top_countries.plot(kind='bar')
plt.title('Top 10 Countries Producing Most Players')
plt.show()
```

### Overall Rating vs Age

```
sns.scatterplot(x=df['Age'], y=df['Overall'])
plt.title('Overall Rating vs Age')
plt.show()
```

### Earnings Comparison Among Offensive Roles

```
offensive_positions = df[df['Position'].isin(['ST','RW','LW'])]
offensive_positions.groupby('Position')['Wage'].mean()
```
